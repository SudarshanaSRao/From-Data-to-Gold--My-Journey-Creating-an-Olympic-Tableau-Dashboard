```python
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
import seaborn as sns

warnings.filterwarnings("ignore")
```

```python
IN_PATH = '/content/drive/MyDrive/Project/Summer-Olympic-medals-1976-to-2008.csv'
OUT_PATH = '/content/drive/MyDrive/Project/Olympics_clean.csv'
```

```python
df = pd.read_csv(IN_PATH, encoding='ISO-8859-1')
df.shape
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True, axis=0)
```

```python
def swap_names(name):
  parts = name.split(', ')
  if len(parts) == 2:
    return f"{parts[1]} {parts[0]}"
  elif len(parts) == 3:
    return f"{parts[1]} {parts[2]}"

  return name

df['Athlete'] = df['Athlete'].apply(swap_names)
df['Athlete'] = df['Athlete'].apply(lambda name: ' '.join(word.capitalize() for word in name.split()))

df['Athlete'].head()
```

```python
df['Year'] = df['Year'].astype(int)
```

```python
df.to_csv(OUT_PATH, index=False)
```
